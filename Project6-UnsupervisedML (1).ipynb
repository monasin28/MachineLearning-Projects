```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("diabetes.csv") #we are using this dataset just to create dendro grams otherwise this dataset is only used for supervised ML because it has both input and output
```

```python
data.head()
```

```python
df=data[['Age','Insulin']]
df.iloc[0:10,:].plot(kind='bar')
```

```python
import scipy.cluster.hierarchy as sch
plt.title("Dendrogram for Diabetes dataset")
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
```

```python
#KMeans clustering
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(3) #here 3 indicates the number of clusters which we have seen in above dendrogram
```

```python
kmeans.fit(df) #we have directly started with fit() function because this is unsupervised ML, we don't need to divide the data which we used to do in supervised ML
```

```python
clusters=kmeans.fit_predict(df)
```
