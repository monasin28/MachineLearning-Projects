```python
#dividing the data into independent and dependent variable, since there is no categoricak data and no null values so no need of preprocessing
x=data.drop(['Outcome'],axis=1) #independent variables
y=data['Outcome'] #dependent variable
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_hat=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_hat)
```
